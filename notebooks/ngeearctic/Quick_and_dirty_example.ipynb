```python
import matplotlib.pyplot as plt
import xarray
import os
```

```python
output_file='~/output/cime_run_dirs/OLMT_AK-K64G_ICB20TRCNPRDCTCBC/run/ELM_output.nc'
```

```python
output=xarray.open_dataset(output_file).squeeze()
```

```python
time = output['time']  ### time dimension, put in unit of years
print(time)

GPP = output['GPP']
print(GPP)
gpp2 = GPP*3600
print(gpp2)

TLAI = output['TLAI']
print(TLAI)

TOTVEGC = output['TOTVEGC']
```

```python
plt.plot(time,TLAI, label='TLAI')
plt.xlabel(''.join(['time (days since 1850 00:00:00)']))
plt.ylabel('LAI (m2/m2)')
```

```python
plt.plot(time,gpp2, label='GPP')
plt.xlabel(''.join(['time (days since 1850 00:00:00)']))
plt.ylabel('GPP (gC/m2/hr)')
```

```python
plt.plot(time,TOTVEGC, label='TOTVEGC')
plt.xlabel(''.join(['time (days since 1850 00:00:00)']))
```
